```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np

W_0 = np.ones((2, 1))
W_0 = np.array([[10], [10]])
b = np.zeros(1)
b = np.array([-10])

model = keras.Sequential(
    [
 #       layers.Dense(1, activation="linear"),
        layers.Dense(1, weights=[W_0, b], input_dim=2, activation="sigmoid"),
    ]
)  # No weights at this stage!

# At this point, you can't do this:
# model.weights

# You also can't do this:
# model.summary()

# Call the model on a test input
x = tf.ones((1, 2))
y = model(x)
print("model paramters:", model.weights)
print("output: ", y)
model.summary()
```
